### Merging BTC files into a single file:

```python

# Path to folder containing BTC CSV files
path = "dataset/BTC/"  # update if your folder name is different

# Get all CSV files
all_files = glob.glob(path + "*.csv")

# Read and concatenate all CSVs using ';' separator
df_list = [pd.read_csv(file, sep=';') for file in all_files]
merged_df = pd.concat(df_list, ignore_index=True)

# Remove any leading/trailing spaces in column names
merged_df.columns = merged_df.columns.str.strip()

# Convert 'timestamp' column to datetime
merged_df['timestamp'] = pd.to_datetime(merged_df['timestamp'])

# Sort data by timestamp
merged_df.sort_values('timestamp', inplace=True)

# Remove duplicates (optional but recommended)
merged_df.drop_duplicates(inplace=True)

# Save the final merged file
merged_df.to_csv("BTC_merged_2010_to_2025.csv", index=False)

print("✅ Merging complete! Saved as BTC_merged_2010_to_2025.csv")
```

### Cleaning and calculating sentiment score of News Datset:

```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# === Step 1: Load and clean news data ===
news_df = pd.read_csv("dataset/News/news_currencies_source_joinedResult.csv")  # Replace with actual file path

# Keep only needed columns
news_df = news_df[['newsDatetime', 'title', 'description', 'currencies']]

# Drop rows with no title or no coin information
news_df.dropna(subset=['title', 'currencies'], inplace=True)

# Fill missing descriptions
news_df['description'] = news_df['description'].fillna('')

# Combine title and description into a new column
news_df['text'] = news_df['title'] + " " + news_df['description']

# Convert datetime and extract date only
news_df['date'] = pd.to_datetime(news_df['newsDatetime']).dt.date
```

```python
analyzer = SentimentIntensityAnalyzer()
news_df['sentiment'] = news_df['text'].apply(lambda x: analyzer.polarity_scores(x)['compound'])
```

```python
# Helper function to filter and aggregate
def get_daily_sentiment(df, coin_symbol):
    df_coin = df[df['currencies'].str.contains(coin_symbol, na=False)].copy()
    daily_sentiment = df_coin.groupby('date')['sentiment'].mean().reset_index()
    daily_sentiment.columns = ['timestamp', 'sentiment']
    return daily_sentiment

# Generate sentiment per coin
btc_sentiment = get_daily_sentiment(news_df, 'BTC')
eth_sentiment = get_daily_sentiment(news_df, 'ETH')
bnb_sentiment = get_daily_sentiment(news_df, 'BNB')
```

```python
btc_sentiment.to_csv("BTC_sentiment.csv", index=False)
eth_sentiment.to_csv("ETH_sentiment.csv", index=False)
bnb_sentiment.to_csv("BNB_sentiment.csv", index=False)

print("✅ Sentiment files saved")
```

### Merging the sentiment column into each coin dataset and saving the final cleaned files ready for model training.

```python
# Example: Merge BTC sentiment into BTC price data
btc_df = pd.read_csv("BTC_cleaned.csv", parse_dates=['timestamp'])
btc_sentiment = pd.read_csv("BTC_sentiment.csv", parse_dates=['timestamp'])

# Merge (keep all price data, add sentiment if available)
btc_merged = pd.merge(btc_df, btc_sentiment, on='timestamp', how='left')

# Fill missing sentiment with 0 (or neutral)
btc_merged['sentiment'] = btc_merged['sentiment'].fillna(0)

```

```python
btc_merged.to_csv("BTC_final.csv", index=False)
```

```python
# Load cleaned ETH price data and ETH sentiment
eth_df = pd.read_csv("ETH_cleaned.csv", parse_dates=['timestamp'])
eth_sentiment = pd.read_csv("ETH_sentiment.csv", parse_dates=['timestamp'])

# Merge sentiment with ETH prices
eth_merged = pd.merge(eth_df, eth_sentiment, on='timestamp', how='left')

# Fill missing sentiment with neutral (0)
eth_merged['sentiment'] = eth_merged['sentiment'].fillna(0)
```

```python
# Load cleaned BNB price data and BNB sentiment
bnb_df = pd.read_csv("BNB_cleaned.csv", parse_dates=['timestamp'])
bnb_sentiment = pd.read_csv("BNB_sentiment.csv", parse_dates=['timestamp'])

# Merge sentiment with BNB prices
bnb_merged = pd.merge(bnb_df, bnb_sentiment, on='timestamp', how='left')

# Fill missing sentiment with neutral (0)
bnb_merged['sentiment'] = bnb_merged['sentiment'].fillna(0)
```

```python
eth_merged.to_csv("ETH_final.csv", index=False)
bnb_merged.to_csv("BNB_final.csv", index=False)
```
